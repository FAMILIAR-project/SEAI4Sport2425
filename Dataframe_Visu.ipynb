Exercice : Trouver et Analyser le Match le Plus Spectaculaire
Question : Quel match peut être considéré comme le plus spectaculaire ? Un match est jugé spectaculaire s'il combine un grand nombre de buts marqués et un faible écart entre les équipes. Identifiez ce match, affichez ses statistiques, et visualisez ses détails.

Solution


```python
# Définir une métrique pour évaluer le "spectacle" d'un match
# Un match spectaculaire combine un grand nombre de buts et un faible écart
data['spectacle_score'] = data['total_goals'] - abs(data['goal_home'] - data['goal_away'])

# Trouver le match avec le score de spectacle le plus élevé
most_spectacular_match = data.loc[data['spectacle_score'].idxmax()]

# Afficher les détails du match
print("Match le plus spectaculaire :")
print(f"{most_spectacular_match['club_home']} vs {most_spectacular_match['club_away']}")
print(f"Score : {most_spectacular_match['goal_home']} - {most_spectacular_match['goal_away']}")
print(f"Spectacle Score : {most_spectacular_match['spectacle_score']}")

# Visualiser le score des deux équipes
teams = [most_spectacular_match['club_home'], most_spectacular_match['club_away']]
goals = [most_spectacular_match['goal_home'], most_spectacular_match['goal_away']]

plt.bar(teams, goals, color=['blue', 'red'])
plt.title(f"Spectacular Match: {teams[0]} vs {teams[1]}")
plt.ylabel('Goals')
plt.show()
```

Explication :

 * Le score de spectacle est calculé comme le nombre total de buts diminué de l'écart entre les équipes, ce qui favorise des matchs serrés mais riches en buts.
 * L'exercice aide les étudiants à explorer les données et à combiner des critères pour analyser des aspects spécifiques d'un match.

Exercice : Trouver le Match le Plus Spectaculaire Basé sur les xG
Question : Quel match est le plus spectaculaire selon les xG (Expected Goals) ? Un match est jugé spectaculaire s'il combine un grand total de xG (attente de buts) et un faible écart entre les xG des deux équipes. Identifiez ce match, affichez ses statistiques et visualisez les xG.

Solution en Python

```python
# Définir une métrique pour évaluer le "spectacle" basé sur les xG
# Un match spectaculaire combine un grand total de xG et un faible écart
data['xG_spectacle_score'] = data['xG'] + data['xGA'] - abs(data['xG'] - data['xGA'])

# Trouver le match avec le score de spectacle le plus élevé basé sur les xG
most_spectacular_xG_match = data.loc[data['xG_spectacle_score'].idxmax()]

# Afficher les détails du match
print("Match le plus spectaculaire basé sur les xG :")
print(f"{most_spectacular_xG_match['club_home']} vs {most_spectacular_xG_match['club_away']}")
print(f"Score : {most_spectacular_xG_match['goal_home']} - {most_spectacular_xG_match['goal_away']}")
print(f"xG Home : {most_spectacular_xG_match['xG']:.2f}")
print(f"xG Away : {most_spectacular_xG_match['xGA']:.2f}")
print(f"xG Spectacle Score : {most_spectacular_xG_match['xG_spectacle_score']:.2f}")

# Visualiser les xG des deux équipes
teams = [most_spectacular_xG_match['club_home'], most_spectacular_xG_match['club_away']]
xG_values = [most_spectacular_xG_match['xG'], most_spectacular_xG_match['xGA']]

plt.bar(teams, xG_values, color=['blue', 'orange'])
plt.title(f"Spectacular Match (xG): {teams[0]} vs {teams[1]}")
plt.ylabel('Expected Goals (xG)')
plt.show()
```

Explication :

* Les xG sont une mesure avancée utilisée en analyse sportive pour estimer le nombre de buts qu'une équipe aurait "dû" marquer.
* Le xG_spectacle_score favorise les matchs avec un total élevé de xG tout en pénalisant les déséquilibres entre les deux équipes.
* Cet exercice montre comment les données avancées (comme les xG) peuvent enrichir l'analyse sportive, tout en utilisant des concepts de programmation.

Exercice : Trouver le Match Avec le Plus d'Occasions Créées Basé sur les xG
Question : Quel match est le plus riche en occasions créées, c'est-à-dire le total des xG (Expected Goals) générés par les deux équipes ? Identifiez ce match, affichez ses statistiques et visualisez les xG des équipes.

Solution en Python

```python
# Calculer le total des xG (occasions créées) pour chaque match
data['total_xG'] = data['xG'] + data['xGA']

# Trouver le match avec le total de xG le plus élevé
most_opportunities_match = data.loc[data['total_xG'].idxmax()]

# Afficher les détails du match
print("Match avec le plus d'occasions créées :")
print(f"{most_opportunities_match['club_home']} vs {most_opportunities_match['club_away']}")
print(f"Score : {most_opportunities_match['goal_home']} - {most_opportunities_match['goal_away']}")
print(f"xG Home : {most_opportunities_match['xG']:.2f}")
print(f"xG Away : {most_opportunities_match['xGA']:.2f}")
print(f"Total xG : {most_opportunities_match['total_xG']:.2f}")

# Visualiser les xG des deux équipes
teams = [most_opportunities_match['club_home'], most_opportunities_match['club_away']]
xG_values = [most_opportunities_match['xG'], most_opportunities_match['xGA']]

plt.bar(teams, xG_values, color=['blue', 'orange'])
plt.title(f"Match With Most Opportunities: {teams[0]} vs {teams[1]}")
plt.ylabel('Expected Goals (xG)')
plt.show()
```

Explication :

* Total xG est une mesure du potentiel offensif des deux équipes pendant un match. Un match avec un total xG élevé indique que les deux équipes ont créé de nombreuses occasions dangereuses.
* Cet exercice est utile pour explorer les matchs les plus offensifs de la saison, en utilisant des statistiques avancées pour l'analyse sportive.

Objectif : Colorer les lignes des matchs où le score est nul (0-0) en gris.

```python
def highlight_goalless(row):
    if row['goal_home'] == 0 and row['goal_away'] == 0:
        return ['background-color: lightgrey;'] * len(row)
    else:
        return [''] * len(row)

styled = data.style.apply(highlight_goalless, axis=1)
styled
```

Objectif : Mettre en gras les équipes ayant marqué au moins 4 buts à domicile.

Colorer les cellules en fonction des scores totaux
Objectif : Appliquer une couleur différente selon si les scores sont faibles (<2), moyens (2-4) ou élevés (>4).

```python
def color_total_goals(value):
    if value < 2:
        return 'background-color: lightblue;'
    elif 2 <= value <= 4:
        return 'background-color: lightgreen;'
    else:
        return 'background-color: orange;'

styled = data[['club_home', 'total_goals']].style.applymap(color_total_goals, subset='total_goals')
styled
```
